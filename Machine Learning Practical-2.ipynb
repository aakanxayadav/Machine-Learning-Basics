```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats


```

```python
#set the random seed for reproducibility
np.random.seed(42)
```

```python
#generate sample data
data= np.random.normal(loc=0, scale=1, size=1000)
```

```python
#create a dataframe
df= pd.DataFrame(data, columns=['Value'])
```

```python
#1. Calculate Gaussian/Normal Distribution
mean = np.mean(data)
std_dev= np.std(data)
print(f"Mean: {mean}, Standard Deviation: {'std_dev'}")
```

```python
#1. plot the guassian/normal distribution
plt.figure(figsize= (10,6))
sns.histplot(data, kde= False, stat="density", bins=30, label='Histogram')
xmin, xmax = plt.xlim()
x = np.linspace(xmin,xmax,100)
p= stats.norm.pdf(x, mean, std_dev)
plt.plot(x, p , 'k', linewidth=2, label='Normal Distribution')
plt.title('Gaussian/Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
#2.Kernal density estimation(KDE)
plt.figure(figsize=(10,6))
sns.kdeplot(data, shade=True, label='KDE')
plt.title('Kernal Density Estimation')
plt.xlabel('Value')
plt.ylabel('Density')
plt.legend()
plt.show()
```

```python
#3. QQ plot
plt.figure(figsize=(10,6))
stats.probplot(data, dist="norm", plot=plt)
plt.title("Q-Q plot")
plt.xlabel('Theoritical Quantiles')
plt.ylabel('Sample Quantile')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from scipy.stats import norm
from sklearn.datasets import load_iris
```

```python
# Load the Iris dataset
iris = load_iris()
sepal_length = iris.data[10, 0]
```

```python
 # Plot the PDF using seaborn
plt.figure(figsize=(10, 6))
```

```python
# PDF Plot
plt.subplot(1, 2, 1)
sns.histplot(sepal_length, kde=True, stat="density", bins=20, color='blue', label='Sepal Length')
plt.title('Probability Density Function (PDF)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Density')
plt.legend()
```

```python
 
# CDF Plot
plt.subplot(1, 2, 2)
sorted_sepal_length = np.sort(sepal_length)
cdf = np.arange(len(sorted_sepal_length)) / float(len(sorted_sepal_length))
```

```python
 
plt.plot(sorted_sepal_length, cdf, color='green', label='Sepal Length')
plt.title('Cumulative Distribution Function (CDF)')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('CDF')
plt.legend()
```

```python
 
# Show plots
plt.tight_layout()
plt.show()
```
